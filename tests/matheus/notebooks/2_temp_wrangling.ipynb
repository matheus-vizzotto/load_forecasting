```python
import pandas as pd
import os
os.listdir()
```

```python
path = "temp_01012000_30062022.csv"
```

```python
df = pd.DataFrame()
chunksize = 10**6
for chunk in pd.read_csv(path, decimal = ",", chunksize=chunksize):
    df1 = chunk[chunk.UF.isin(["PR", "SC", "RS"])]
    df = pd.concat([df, df1])
```

```python
for col in df.columns:
    print(col)
```

```python
df.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
# AJUSTANDO NOME DE COLUNAS

new_cols = []
for col in df.columns:
    splitted = col.split(" (")
    name = splitted[0]
    adj = name.replace(".", "").replace(",", "").replace("-", "").replace(" ", "_")
    new_cols.append(adj)

df.columns = new_cols

for col in df.columns:
    print(col)
```

Unificar colunas: <br>
&ensp;&ensp;&ensp;DATA e Data <br>
&ensp;&ensp;&ensp;HORA(UTC) e Hora UTC <br>
&ensp;&ensp;&ensp;RADIACAO GLOBAL (KJ/m²) e RADIACAO GLOBAL (Kj/m²)

```python
df["DATA"] = df["DATA"].combine_first(df["Data"])
df["HORA"] = df["HORA"].combine_first(df["Hora_UTC"])
df.iloc[:, 6] = df.iloc[:, 6].combine_first(df.iloc[:, -1])
```

```python
df = df.iloc[:, :-3]
```

```python
df.drop("Unnamed:_19", axis = 1, inplace = True)
```

A coluna [DATA] aparenta ter dois formatos: '%Y-%m-%d' e '%Y/%m/%d' <br>
A coluna [HORA] também: '%H:%M' e '%H%M UTC'

```python
def ajusta_hora(x):
    if ":" in x:
        y = x
    elif "UTC" in x:
        y = x[:2] + ":" + x[2:4]
    else:
        y = None
    return y

def converte_data(x):
    try:
        y = pd.to_datetime(x, format = '%Y-%m-%d %H:%M:%S')
    except:
        y = pd.to_datetime(x, format = '%Y/%m/%d %H:%M:%S')
    return y
```

```python
df.loc[:, "HORA"] = df.loc[:, "HORA"].apply(ajusta_hora)
df["DATA_HORA"] = df["DATA"] + " " + df["HORA"]
df.loc[:, "DATA_HORA"] = df.loc[:, "DATA_HORA"].apply(converte_data) 
```

```python
df.loc[:, "VENTO_RAJADA_MAXIMA"] = df.loc[:, "VENTO_RAJADA_MAXIMA"].apply(lambda x: float(str(x).replace(",",".")))
df.loc[:, "VENTO_VELOCIDADE_HORARIA"] = df.loc[:, "VENTO_VELOCIDADE_HORARIA"].apply(lambda x: float(str(x).replace(",",".")))
df.loc[:, "TEMPERATURA_ORVALHO_MIN_NA_HORA_ANT"] = df.loc[:, "TEMPERATURA_ORVALHO_MIN_NA_HORA_ANT"].apply(lambda x: float(str(x).replace(",",".")))
df.loc[:, "RADIACAO_GLOBAL"] = df.loc[:, "RADIACAO_GLOBAL"].apply(lambda x: float(str(x).replace(",",".")))
df.loc[:, "PRECIPITAÇÃO_TOTAL_HORÁRIO"] = df.loc[:, "PRECIPITAÇÃO_TOTAL_HORÁRIO"].apply(lambda x: float(str(x).replace(",",".")))
```
