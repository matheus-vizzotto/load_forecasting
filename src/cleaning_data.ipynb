```python
def merge_dataframes(data1: pd.DataFrame, data2: pd.DataFrame, date1: str, date2: str, dt_min, dt_max, freq_='h', reset_index=False) -> pd.DataFrame:
    dt_range = pd.date_range(dt_min, dt_max, freq=freq_).to_series(name="datetime")
    df1 = pd.merge(dt_range, data1, left_on = "datetime", right_on=date1, how='outer')
    df2 = pd.merge(df1, data2, left_on=date1, right_on=date2, how='outer')
    return df2
```

```python
merge_dataframes(data1=df_load.reset_index(), data2=df_temp.reset_index(), date1="date", date2="data_hora", dt_min=df_load.index.min(), dt_max=df_load.index.max())
```
